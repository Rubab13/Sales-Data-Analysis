<div style='background-color:#f4d8cd;'>
    <h2 style='color:#1e555c;font-weight:bold;font-size:1.5rem;padding:1rem;'>Is there any correlation between product price and sales volume?</h2>
</div>

#### Investigate whether higher-priced products tend to sell in larger quantities.

```python
correlation = np.corrcoef(data['PRICEEACH'], data['QUANTITYORDERED'])[0, 1]
print(f"Correlation coefficient between PRICEEACH and QUANTITYORDERED: {correlation}")
```

#### A very weak positive correlation
Since the correlation coefficient is close to zero, it suggests that there is almost no linear relationship between the price of the product and the quantity ordered. In other words, knowing the price of a product does not provide much information about how many units will be sold.
A positive coefficient indicates that as one variable (price) increases, the other variable (quantity ordered) tends to increase slightly, but the relationship is so weak that it is practically negligible.

```python
lm = sns.lmplot(x='PRICEEACH', y='QUANTITYORDERED', data=data, aspect=1.5, scatter_kws={'color': '#000', 'alpha': 0.7}, line_kws={'color': 'red'})
lm.set_axis_labels('Price Each', 'Quantity Ordered')
lm.fig.suptitle('Relationship between Price and Sales Volume', y=1.03)
plt.show()
```

<div style='background-color:#EAD7D7;'>
    <p style='color:#1e555c;font-weight:400;font-size:1.3rem;padding:1rem'>The best fit line is horizontal. Almost parallel to the x-axis, which predicts that we cannot make an assumption on the basis of price about the sales. However, a very slight assumption based on the positive sign is this that with the increase of price there is a 1% chance is that sales will increase. The result is people tend to buy cheaper products.</p>
</div>

    <h2 style='color:#1e555c;font-weight:bold;font-size:1.5rem;padding:1rem;'>Are there regional differences in product preferences?</h2>
</div>

```python
country_product_counts = data.groupby(['COUNTRY', 'PRODUCTLINE']).size().reset_index(name='COUNT')
country_product_pivot = country_product_counts.pivot(index='COUNTRY', columns='PRODUCTLINE', values='COUNT').fillna(0)
plt.figure(figsize=(10, 8))
sns.heatmap(country_product_pivot, annot=True, cmap='viridis', fmt='.0f')
plt.title('Product Preferences Across Countries')
plt.xlabel('Product Line')
plt.ylabel('Country')
plt.show()
```

#### Lighter shades indicate higher counts, showing which product lines are more popular in specific countries.<br>
#### Darker shades represent lower counts, indicating less popularity.<br>
#### The numbers on the heatmap represent the exact count of each product line within each country.

<div style='background-color:#EAD7D7;'>

<div style='background-color:#f4d8cd;'>
    <h2 style='color:#1e555c;font-weight:bold;font-size:1.5rem;padding:1rem;'>How does the MSRP (Manufacturer's Suggested Retail Price) compare with actual PRICEEACH?</h2>
</div>

```python
data['PRICE_DIFF'] = data['PRICEEACH'] - data['MSRP']
data['PRICE_DIFF_PERCENT'] = ((data['PRICEEACH'] - data['MSRP']) / data['MSRP']) * 100
plt.figure(figsize=(8, 6))
sns.scatterplot(x='MSRP', y='PRICEEACH', data=data, color='b', s=100, alpha=0.8)
plt.plot(data['MSRP'], data['MSRP'], color='r', label='MSRP = PRICEEACH')
plt.title('Comparison of MSRP vs PRICEEACH')
plt.xlabel('MSRP (Manufacturer\'s Suggested Retail Price)')
plt.ylabel('PRICEEACH (Actual Price)')
plt.legend()
plt.show()
```
